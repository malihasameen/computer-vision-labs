# **Lab06: Hough Transform**

```python
# Import Statements
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.feature import canny
from skimage.transform import hough_circle, hough_circle_peaks, probabilistic_hough_line
from skimage import color
import skimage
import numpy as np
```

```python
image = imread("lab_06 (1).jpg", as_gray=True)
result_image = color.gray2rgb(image)
edges = canny(image, 2, 1, 25)

# Detect Lines
lines = probabilistic_hough_line(edges, threshold=5, line_length=1,line_gap=1)
for line in lines:
  p0, p1 = line
  rr, cc = skimage.draw.line(p0[1], p0[0], p1[1], p1[0])
  result_image[rr, cc] = (0, 0, 255)

# Detect Circles
hough_radii = np.arange(25, 50)
hough_res = hough_circle(edges, hough_radii)
accums, cx, cy, radii = hough_circle_peaks(hough_res, hough_radii,
                                           total_num_peaks=2)

for center_y, center_x, radius in zip(cy, cx, radii):
    circy, circx = skimage.draw.circle_perimeter(center_y, center_x, radius,
                                    shape=result_image.shape)
    result_image[circy, circx] = (255, 0, 0)

plt.imshow(result_image)
```
