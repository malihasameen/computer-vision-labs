```python
import skimage
from skimage.transform import pyramid_gaussian, pyramid_laplacian
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = skimage.data.chelsea()
plt.imshow(img)
```

## **Gaussian Pyramid**

```python
pyd = tuple(pyramid_gaussian(img, downscale=3, multichannel=True))

c_image = np.zeros((img.shape[0], img.shape[1] + img.shape[1] // 2, 3), dtype=np.double)
c_image[:img.shape[0], :img.shape[1], :] = pyd[0]

i = 0
for p in pyd[1:]:
    n_rows, n_cols = p.shape[:2]
    c_image[i:i + n_rows, img.shape[1]:img.shape[1] + n_cols] = p
    i += n_rows

fig=plt.figure(figsize=(10,8))
fig.add_subplot(1, 1, 1, xticks=[], yticks=[])
plt.title("Gaussian Pyramid")
plt.imshow(c_image)
```
